# II Solving The Maximum Pairwise Product Programming Challenge in C++

## C++

$$ T(n) < 6 n^2 = O(n^2)$$

````python
%%writefile max_pairwise_product_full.cpp
#include <iostream>
#include <vector>
#include <algorithm>

int MaxPairwiseProduct(const std::vector<int>& numbers) {
    int max_product = 0;
    int aux_product = 0;
    int n = numbers.size();

    for (int first = 0; first < n; ++first) {
        for (int second = 0; second < n; ++second) {
          if ( first != second ){
                aux_product = numbers[first] * numbers[second];
                if (aux_product > max_product)  max_product = aux_product;
          }
        }
    }

    return max_product;
}

int main() {
    int n;
    std::cin >> n;
    std::vector<int> numbers(n);
    for (int i = 0; i < n; ++i) {
        std::cin >> numbers[i];
    }

    std::cout << MaxPairwiseProduct(numbers) << "\n";
    return 0;
}
````

#### Time Complexity
$$\left. \begin{array}{llc}
 \text{INSTRUCTION} & \text{TYPE}& \text{# EXECUTIONS}\\
\text{first != second} &  \text{comparisons} &n^2 \\
\text{numbers[first] * numbers[second]}& \ \text{ products}  &n^2 - n \\
\text{max_product = aux_product} &  \text{assigments} & [1, \frac{n^2 - n}{2})  
\end{array}\right\}  =  O(n^2)$$

### Calculate the number or and upper bound to the number of instructions executed by the algorithm as a function of n, add in each line a comment with the number of instructions that are executed in that line.

````python
%%writefile max_pairwise_product_long_full.cpp
#include <iostream>
#include <vector>
#include <algorithm>

long long MaxPairwiseProduct(const std::vector<int>& numbers) {              //INSTRUCTIONS executed by line
    long long max_product = 0LL;                                             //1
	  long long aux_product = 0LL;                                             //1
    int n = numbers.size();                                                  //1

    for (int first = 0; first < n; ++first) {                                //2+2n
        for (int second = 0; second < n; ++second) {                         //2n + 2n**2
          if ( first != second ){                                            //n**2
                aux_product = numbers[first] * numbers[second];              //n**2 -n
                if (aux_product > max_product)                               //n**2 -n
                    max_product = aux_product;                               //[0, n**2/2 - n/2] upper bound
          }
        }
    }

                                                                             // --------------
    return max_product;                                                      // 5 + 2*n + [5*n**2, 11*n**2/2  - n/2]  upper bound
}

int main() {                                              //INSTRUCTIONS executed by line
    int n;                                                //1
    std::cin >> n;                                        //1
    std::vector<int> numbers(n);                          //n
    for (int i = 0; i < n; ++i) {                         //2 + 2n
        std::cin >> numbers[i];                           //n
    }

    std::cout << MaxPairwiseProduct(numbers) << "\n";     //1 + 1  + 5 + 2*n + [5*n**2, 11*n**2/2  - n/2]

    return 0;                                             // --------------
}                                                         //  11 + 6*n + [5*n**2, 11*n**2/2  - n/2]  upper bound
````

## Solutions in C++ and Python that uses
$$\frac{n^2 - n}{2} = O(n^2)\ \ \text{products}$$

````python
%%writefile max_pairwise_product.cpp
#include <iostream>
#include <vector>
#include <algorithm>

int MaxPairwiseProduct(const std::vector<int>& numbers) {
    int max_product = 0;
    int n = numbers.size();

    for (int first = 0; first < n; ++first) {
        for (int second = first + 1; second < n; ++second) {
            max_product = std::max(max_product,
                numbers[first] * numbers[second]);
        }
    }

    return max_product;
}

int main() {
    int n;
    std::cin >> n;
    std::vector<int> numbers(n);
    for (int i = 0; i < n; ++i) {
        std::cin >> numbers[i];
    }

    std::cout << MaxPairwiseProduct(numbers) << "\n";
    return 0;
}
````

````python
%%writefile max_pairwise_product_long.cpp
#include <iostream>
#include <vector>
#include <algorithm>

long long MaxPairwiseProduct(const std::vector<int>& numbers) {              //INSTRUCTIONS executed by line
    long long max_product = 0LL;                                             //1
	  long long aux_product = 0LL;                                             //1
    int n = numbers.size();                                                  //1

    for (int first = 0; first < n; ++first) {                                //2+2n
        for (int second = first + 1; second < n; ++second) {                 //2n + (n-1)n
			       aux_product =  (long long) numbers[first] * numbers[second];    //(n-1)n/2
            if (max_product < aux_product)  {                                //(n-1)n/2
				        max_product = aux_product;                                   //[1,(n-1)n/2)  upper bound
             }
        }
    }
                                                                             // --------------
    return max_product;                                                      // 5 + 4n + [2(n-1)n+1, 2.5 (n-1)n) upper bound
}

int main() {                                              //INSTRUCTIONS executed by line
    int n;                                                //1
    std::cin >> n;                                        //1
    std::vector<int> numbers(n);                          //n
    for (int i = 0; i < n; ++i) {                         //2 + 2n
        std::cin >> numbers[i];                           //n
    }

    std::cout << MaxPairwiseProduct(numbers) << "\n";     //1 + 1  + 5 + 4n + [2(n-1)n+1, 2.5 (n-1)n) upper bound
    return 0;                                             // --------------
}                                                         // 11 + 7n + [2(n-1)n+1, 2.5 (n-1)n) upper bound
````

**Introsort**
https://en.wikipedia.org/wiki/Introsort

Introsort or introspective sort is a hybrid sorting algorithm that provides both fast average performance and (asymptotically) optimal worst-case performance. It begins with quicksort, it switches to heapsort when the recursion depth exceeds a level based on (the logarithm of) the number of elements being sorted and it switches to insertion sort when the number of elements is below some threshold

*   Class	Sorting algorithm
*   Data structure	Array
*   Worst-case performance	$O(n \lg n)$
*   Average performance	$O(n \lg n)$

**Tinsort**

Timsort is a hybrid, stable sorting algorithm, derived from merge sort and insertion sort, designed to perform well on many kinds of real-world data. It was implemented by Tim Peters in 2002 for use in the Python programming language.

*   Data structure	Array

*   Best-case performance $O(n)$

*   Worst-case space complexity $O(n)$

````python
%%writefile max_pairwise_product_sort_cpp.cpp
#include <iostream>
#include <vector>
#include <algorithm>

using namespace std;

long long maxPairwiseProduct(const vector<int>& arr) {
    vector<int> sortedArr = arr;
    sort(sortedArr.begin(), sortedArr.end());
    long long max1 = static_cast<long long>(sortedArr[sortedArr.size() - 1]);
    long long max2 = static_cast<long long>(sortedArr[sortedArr.size() - 2]);

    return max1 * max2;
}

int main() {
    int n;
    cin >> n;

    vector<int> arr(n);
    for (int i = 0; i < n; ++i) {
        cin >> arr[i];
    }

    long long result = maxPairwiseProduct(arr);
    cout << result;
    return 0;
}
````

````python
%%writefile max_pairwise_product_efficient.cpp
#include <iostream>
#include <vector>
using namespace std;

long long maxPairwiseProduct(const vector<int>& arr) {
    int n = arr.size(); // 1 instrucción
    int max1 = -1, max2 = -1; // 2 instrucciones

    // Encontrar el mayor elemento
    for (int i = 0; i < n; ++i) { // n instrucciones para el bucle
        if (max1 == -1 || arr[i] > arr[max1]) { // 2n comparaciones
            max2 = max1; // n asignaciones
            max1 = i; // n asignaciones
        } else if (max2 == -1 || arr[i] > arr[max2]) { // 2n comparaciones
            max2 = i; // n asignaciones
        }
    }

    return static_cast<long long>(arr[max1]) * arr[max2]; // 1 multiplicación y retorno
}

int main() {
    int n;
    cin >> n; // 1 lectura

    vector<int> arr(n); // 1 inicialización
    for (int i = 0; i < n; ++i) { // n asignaciones
        cin >> arr[i]; // n lecturas
    }

    long long result = maxPairwiseProduct(arr); // 1 llamada de función
    cout << result; // 1 escritura
    return 0;
}
````

````python
import subprocess
import random

def efficientSolution(numbers):
    largest = 0
    secondLargest = 0
    for i in numbers:
        if i > largest:
            secondLargest = largest
            largest = i
        elif i > secondLargest:
            secondLargest = i
    return secondLargest * largest

def generate_test_case(max_size=10000, max_value=10000000):
    size = random.randint(2, max_size)
    numbers = [random.randint(0, max_value) for _ in range(size)]
    return size, numbers

for test_id in range(20):
    size, numbers = generate_test_case()
    input_data = f"{size}\n{' '.join(map(str, numbers))}\n"

    # Mostrar el input data
    print(f"Test {test_id + 1}:")
    print(f"Input data: {input_data.strip()}")

    # Ejecutar el código Python y capturar salida
    py_result = subprocess.run(
        ["python3", "max_pairwise_product_efficient.py"],
        input=input_data,
        text=True,
        capture_output=True
    ).stdout.strip()

    efficient = efficientSolution(numbers)

    # Validar que los resultados sean iguales
    if int(py_result) != efficient:
        print(f"❌ Test {test_id + 1} failed!")
        print(f"Python Output: {py_result}")
        print(f"Output efficient: {efficient}\n")
        break
    else:
        print(f"✅ Test {test_id + 1} passed!")
        print(f"Output python: {py_result}\n")
````

## Perform 100 randomized automated tests in C++ and Python with a porgram  compares the answers,for the same 100 random input instances, of the super efficient solution $(T(n) < 10 n = O(n))$ against the sorting solution  $(T(n)  = O(n \lg{n}))$,  if the answers are not the same stops and displays "Error" listing  the input and the two answers; if the 100 tests are completed without stoping displays "Pass".

````python
import subprocess
import random

!g++ -o max_pairwise_product_sort_cpp max_pairwise_product_sort_cpp.cpp
!g++ -o max_pairwise_product_super_efficient max_pairwise_product_super_efficient.cpp

def generate_test_case(max_size=10000, max_value=10000000):
    size = random.randint(2, max_size)
    numbers = [random.randint(0, max_value) for _ in range(size)]
    return size, numbers

for test_id in range(100):
    size, numbers = generate_test_case()
    input_data = f"{size}\n{' '.join(map(str, numbers))}\n"

    # Ejecutar el código C++ y capturar salida
    cpp_result = subprocess.run(
        ["./max_pairwise_product_sort_cpp"],
        input=input_data,
        text=True,
        capture_output=True
    ).stdout.strip()

    cpp_result_efficient = subprocess.run(
        ["./max_pairwise_product_super_efficient"],
        input=input_data,
        text=True,
        capture_output=True
    ).stdout.strip()

    # Ejecutar el código Python y capturar salida
    py_result = subprocess.run(
        ["python3", "max_pairwise_product_sort.py"],
        input=input_data,
        text=True,
        capture_output=True
    ).stdout.strip()

    # Ejecutar el código Python y capturar salida
    py_result_efficient = subprocess.run(
        ["python3", "max_pairwise_product_super_efficient.py"],
        input=input_data,
        text=True,
        capture_output=True
    ).stdout.strip()


    # Validar que los resultados sean iguales
    if int(cpp_result) != int(cpp_result_efficient) or int(py_result) != int(py_result_efficient):
        print(f"❌ Test {test_id + 1} failed!")
        print(f"Test {test_id + 1}:")
        print(f"Input data: {input_data.strip()}")
        print(f"C++ Output: {cpp_result}")
        print(f"Python Output: {py_result}")
        print(f"Python Output efficient: {py_result_efficient}")
        print(f"C++ Output efficient: {cpp_result_efficient}\n")
        break

print("✅ All tests passed!")
````
